```python
import matplotlib.pyplot as plt
from basic_func import get_spectral, normalization, moving_avg, stacking
```

```python
def fig():
    plt.figure(figsize=(6,4))##, dpi=70)
def plt_bf():
    plt.legend()
    plt.grid()
    plt.xlabel('Wavelength [Âµm]')
    plt.ylabel('Reflentance')
```

```python
ref = get_spectral("frt00003621_07_if166l_E_-97.59028_N_24.92442.csv")
target = get_spectral("frt00003621_07_if166l_E_-97.69471_N_24.88780.csv")
# ref = get_spectral("I52359002SNU_E_-73.31904_N_30.72080.csv")
# target = get_spectral("I52359002SNU_E_-73.36628_N_30.27659.csv")
# fig()
plt.figure()
plt.plot(ref[0],ref[1],label='Reference Spectrum')
plt.plot(target[0],target[1],label='Target Spectrum')
plt_bf()
plt.title('Spectral')
plt.savefig('CRISM_plot.pdf',bbox_inches='tight')
```

```python
# Normalization
ans = normalization(ref,target)
# fig()
plt.figure()
plt.plot(ans[0], ans[1], label='Normalization')
plt_bf()
plt.ylabel('')
plt.title('Normalization')
plt.savefig('CRISM_Normalization.pdf')
```

```python
# Moving Average
ans = moving_avg(target, 3)

# fig()
plt.figure()
plt.plot(ans[0],ans[1],label='Target Spectrum')
plt.plot(ans[2],ans[3],label='Moving Average')
plt_bf()
plt.title('Moving Average')
plt.savefig('CRISM_MovingAverage.pdf',bbox_inches='tight')
```

```python
# Stacking
target1 = get_spectral("frt00003621_07_if166l_E_-97.60818_N_24.87141.csv")
target2 = get_spectral("frt00003621_07_if166l_E_-97.69471_N_24.88780.csv")
target3 = get_spectral("frt00003621_07_if166l_E_-97.63953_N_24.90938.csv")
# target1 = get_spectral("I52359002SNU_E_-73.36628_N_30.27659.csv")
# target2 = get_spectral("I52359002SNU_E_-73.40203_N_30.30469.csv")
# target3 = get_spectral("I52359002SNU_E_-73.40138_N_30.30338.csv")
ans = stacking(target1,target2,target3)
# fig()
plt.figure()
plt.plot(target1[0],target1[1],label='Target Spectrum 1',lw=0.8)
plt.plot(target2[0],target2[1],label='Target Spectrum 2',lw=0.8)
plt.plot(target3[0],target3[1],label='Target Spectrum 3',lw=0.8)
plt.plot(ans[0],ans[1],label='Stacking')
plt_bf()
plt.title('Stacking')
plt.savefig('CRISM_Stacking.pdf',bbox_inches='tight')
```
